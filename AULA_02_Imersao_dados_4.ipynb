# Criar uma escala de valores em milhões

```python
dados_vendas.head()
```

```python
nova_escala = dados_vendas[dados_vendas['Tipo_anuncio'].isnull()]
nova_escala['Valor_anuncio_float'].apply(lambda x: x/1000000)
```

```python
nova_escala['Valor_anuncio_float'] = nova_escala['Valor_anuncio_float'].apply(lambda x: x/1000000)
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
nova_escala[['Rua','Bairro','Cidade','Metragem','Quartos','Banheiros','Vagas','Moeda','Valor_anuncio_float']].head(10)
```

```python
nova_escala.describe()
```

# Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

plt.figure(figsize=(17,7))
sns.color_palette("pastel")

# valores = sns.load_dataset(nova_escala).sort_values("Valor_anuncio_float", ascending=False)

plot = sns.histplot(data=nova_escala, x="Valor_anuncio_float", kde=True, element="poly")

plt.suptitle('Histograma sobre os Valores dos Imóveis de São Paulo')
plt.title('Valores entre 0 e 15 milhões')
plt.xlabel('Valor do Imóvel')
plt.ylabel('Quantidade de Imóveis à Venda')

plot.legend(['Valores em escala de milhão'], prop={'size': 20})

plt.xlim((0, 15))

plt.show()
```

# Preço do metro quadrado por bairro e plotar em um gráfico ideal

```python
nova_escala[["Bairro", "Metragem","Valor_anuncio_float"]].groupby("Bairro", as_index=False).mean()
preco_metro = nova_escala[["Bairro", "Metragem","Valor_anuncio_float"]].groupby("Bairro", as_index=False).mean()
preco_metro.head()
```

```python
preco_metro["Preco_Metro2"] = preco_metro["Valor_anuncio_float"]/preco_metro["Metragem"].apply(lambda x: x/1000)

preco_metro.head()
```

```python
preco_metro_medio = preco_metro.groupby('Bairro').mean()["Valor_anuncio_float"]*1000
preco_medio_bairro = preco_metro.groupby('Bairro').mean()['Metragem']

plt.figure(figsize=(17,8))
plt.ylim((0,15))
(preco_metro_medio/preco_medio_bairro).plot.line()
```

```python
plt.figure(figsize=(20,8))
plt.boxplot(preco_metro['Valor_anuncio_float'], vert=False)
```

```python
plt.figure(figsize=(20,8))
plt.boxplot(preco_metro["Preco_Metro2"], vert=False)
```
